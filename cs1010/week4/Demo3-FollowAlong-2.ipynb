```python
A = np.array([[1, 2, 3], [-1, -2, -3], [6, 5, -5]])
print(A)
```

```python
df_A = pd.DataFrame(A)
print(df_A)
```

```python
column_names = ['Col A', 'Col B', 'Col C']
row_names = ['Happy', 'Halloween', 'Everyone!']

df_A.columns = column_names
df_A.index = row_names

display(df_A)
```

```python
print(df_A.loc['Everyone!', 'Col A'])
```

```python
display(df_A.loc[:, 'Col B'])
```

```python
display(df_A.loc['Halloween', :])
```

```python
arr = df_A.loc['Halloween':, 'Col B' : 'Col C'].values
print(arr)
```

```python
df_A.loc['Everyone!', 'Col B'] = 12
display(df_A)
```
